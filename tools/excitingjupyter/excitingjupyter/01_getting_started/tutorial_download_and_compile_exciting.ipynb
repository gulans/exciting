### <span style="color:#15317E">Introduction</span>

First, you should choose the directory where you want to unpack the current release, <code><strong>exciting <em>fluorine</em></strong></code>. All examples in this tutorial assume that the source code of **`exciting`** is unpacked to  <code>/home/tutorials</code>. From now on the symbol **$** will indicate the shell prompt.

```bash
$ mkdir /home/tutorials 
$ cd /home/tutorials
```

Download the current release, <code><strong>exciting <em>fluorine</em></strong></code> [**here**](http://exciting.wikidot.com/fluorine), and unpack it.

```bash
$ tar xvf exciting.fluorine.tar.gz
```

The latest version of exciting fluorine is also available from exciting's Github page:
<hr style="border:1px solid #DDD"> </hr>

<details>
    
<summary><strong><span style="color:firebrick">$\Rightarrow$ clone from Github</span></strong></summary>

    
```bash
$ git clone https://github.com/exciting/exciting.git
```
</details>
<hr style="border:1px solid #DDD"> </hr>

Before proceeding further, be sure that relevant environment variables are already defined as specified in [**How to set environment variables for tutorials scripts**](http://exciting.wikidot.com/fluorine-tutorial-scripts-and-environment-variables). If you choose a different location than `/home/tutorials`, replace `/home/tutorials` in your `~/.bashrc` file with an appropriate path as indicated in [**How to set environment variables for tutorials scripts**](http://exciting.wikidot.com/fluorine-tutorial-scripts-and-environment-variables).

From now on, we assume that you have already defined all environment variables, including **$EXCITINGROOT**.

Before you compile the code, please also ensure that you have a fortran compiler available. **`exciting`** supports both **<span style="color:mediumblue">GCC</span>** and **<span style="color:mediumblue">Intel</span>** fortran compilers. One can inspect whether a compiler is in the **$PATH** using the `which` command:
<hr style="border:1px solid #DDD"> </hr>

#### <span style="color:#15317E">Inspecting Intel ifort:</span>

```bash
$ which ifort         # next line output is only an example...
/usr/global/intel-2019/compilers_and_libraries_2019.0.117/linux/bin/intel64/ifort
```
<hr style="border:1px solid #DDD"> </hr>

#### <span style="color:#15317E">Inspecting GCC gfortran:</span>

```bash
$ which gfortran      # next line output is only an example...
/usr/bin/gfortran
```
<hr style="border:1px solid #DDD"> </hr>

For detailed information on installing and compiling **`exciting`**, we strongly suggest to read the file **$EXCITINGROOT/INSTALL**.

Assuming a compiler is found, move inside the **$EXCITINGROOT** directory and compile using either of the supported compilers. For the purpose of using exciting within this and the other tutorials, build with **<span style="color:mediumblue">Intel ifort</span>** or **<span style="color:mediumblue">GCC gfortran</span>**, as shown in the following:
<hr style="border:1px solid #DDD"> </hr>

#### <span style="color:#15317E">Compile with Intel ifort:</span>

```bash
$ cd $EXCITINGROOT
$ cp build/platforms/make.inc.ifort build/make.inc
$ make smp
```
<hr style="border:1px solid #DDD"> </hr>

#### <span style="color:#15317E">Compile with GCC gfortran:</span>

```bash
$ cd $EXCITINGROOT
$ cp build/platforms/make.inc.ifort build/make.inc
$ make smp
```
<hr style="border:1px solid #DDD"> </hr>

where the command `make smp` builds the **<span style="color:firebrick">shared-memory multiprocessing version</span>** of the code. If the compilation step was successful, the binary file **exciting_smp** should be available in the **$EXCITINGROOT/bin** directory.

```bash
$ ls $EXCITINGROOT/bin
exciting_smp
```
<hr style="border:1px solid #DDD"> </hr>

If compilation fails, manual editing of the file **make.inc** inside **$EXCITINGROOT/build** may be required. 

However, this is only likely if building with a compiler or libraries installed in non-standard locations. Additional details can be found in the file **$EXCITINGROOT/INSTALL**.

<hr style="border:1px solid #DDD"> </hr>
